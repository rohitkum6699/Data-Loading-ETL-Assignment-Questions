Q8. Cleaning Strategy (Step-by-Step)
Step 1: Remove Duplicates

DELETE FROM table
WHERE Order_ID IN (
  SELECT Order_ID
  FROM table
  GROUP BY Order_ID
  HAVING COUNT(*) > 1

Step 2: Handle Missing Values

For O102:

Options:

1. Replace with average

2. Replace with median

3. Ask business team

UPDATE table
